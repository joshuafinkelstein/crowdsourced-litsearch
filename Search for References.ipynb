```python
# set_of_references = [
#     'https://doi.org/10.1042/BST20120098', 
#     'https://doi.org/10.1016/j.gca.2008.07.016',
#     'https://doi.org/10.1038/s41579-019-0173-x',
#     'https://doi.org/10.3389/fmicb.2015.00575',
#     'https://doi.org/10.1126/science.1066771',
#     'https://doi.org/10.1111/j.1462-2920.2004.00593.x',
#     'https://doi.org/10.1007/s10295-009-0609-9',
#     'https://doi.org/10.1038/srep06961',
#     'https://doi.org/10.1128/AEM.03034-10'
    
#                     ]


starting_article = 'https://doi.org/10.1038/srep45309'
refs = cr.works(ids = starting_article)['message']['reference']
lst_refs = []
for ref in refs:
    try:
        lst_refs += [ref['DOI']]
    except:
        pass

# set of references from starting article
set_of_references = lst_refs

# find references in common
all_references = []
hits = []
for article in set_of_references:
    try:
        refs = cr.works(ids = article)['message']['reference']
    except:
        print('Not available on crossref: ' + article)
#     print(refs)
  
    # add references to the master list
    for ref in refs:
        
        try:
            doi = ref['DOI']        
            all_references += [doi]
        except:
            print('DOI not listed from article: ' + article + '  Ref KEY: ' + ref['key'])

# look for multiple copies
recommendations = {}
for ref in all_references:
    
    hits = 0
    for other_ref in all_references:
        if other_ref == ref:
            hits += 1
            
    if hits > 1:
#         print(ref)
#         print(hits)
        recommendations[ref]= hits
#         recommendations[ref]['link'] = 'https://doi.org/' + ref

# sort by num of hits
recommendations = {k: v for k, v in reversed(sorted(recommendations.items(), key=lambda item: item[1]))}
```

```python
# print links for references to peruse
for rec in recommendations:
    data = cr.works(ids = rec)['message']
    try:
        print(str(data['title'][0]))
        print('Article Hits: ' + str(recommendations[rec]))
        print('Cited By: ' + str(data['is-referenced-by-count']))
    except:
        pass
    
    print('Link: ' + 'https://doi.org/' + rec)
    print('\n')
    
recommendations
```
